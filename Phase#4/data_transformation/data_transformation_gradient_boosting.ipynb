### Handling categorical attributes through e.g., one-hot encoding or conversion to ordinal data

```python
# - handling categorical attributes through e.g., one-hot encoding or conversion to ordinal data,
def oneHotEncode(df_copy): # the last column won't be encoded and only categorical data will be encoded
    
    df = df_copy.copy()
    # remove target col
    target_df = df.pop('age_group')
    
    non_numeric_cols = list(df.select_dtypes([object]).columns)
    
    df_to_encode = df[non_numeric_cols]
    
    cols_not_to_encode = []
    for col in list(df.columns):
        if col not in non_numeric_cols:
            cols_not_to_encode.append(col)
    

    enc = OneHotEncoder(handle_unknown='ignore')
    enc.fit(df_to_encode)
    df_encoded = enc.fit_transform(df_to_encode)
    old_features = list(df_to_encode.columns.values)
    new_features = enc.get_feature_names(old_features)
    df_encoded = pd.DataFrame(df_encoded.toarray(), columns = new_features)
    
    for col in cols_not_to_encode:
        df_encoded[col] = df[col]
        
    # re-adding target col
    df_encoded['age_group'] = target_df
    df_encoded = df_encoded.dropna()
    
    return df_encoded

combined_df_encoded = oneHotEncode(combined_df)
# combined_df_encoded.info()
# combined_df_encoded[combined_df_encoded.isnull().any(axis=1)]
```

### Normalisation of numeric attributes to ensure all attributes are of equal importance during learning

```python
# - normalisation of numeric attributes to ensure all attributes are of equal importance during learning,
def normalize(df_copy):
    
    df = df_copy.copy()
    # remove target col
    target_df = df.pop('age_group')
    

    # normalize
    # get numeric cols
    numeric_cols = list(df.select_dtypes([np.int64,np.float64]).columns)
    
    cols_to_not_normalize = []
    cols_to_normalize = []
    for col in numeric_cols:
        if col.endswith("_id"):
            cols_to_not_normalize.append(col)
        else:
            cols_to_normalize.append(col)
    
    df_not_normalized = df[cols_to_not_normalize]
    df_normalized = df[cols_to_normalize]
    x = df_normalized.values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df_normalized = pd.DataFrame(x_scaled, columns = cols_to_normalize)

    # add non_numeric cols to df_normalized
    non_numeric_cols = list(df.select_dtypes([object]).columns)
    
    for i in non_numeric_cols:
        df_temp = df.pop(i)
        df_normalized[i] = df_temp
        
    #df_non_numeric = df[]
    # re-adding target col
    for col in cols_to_not_normalize:
        df_normalized[col] = df_not_normalized[col]
        
    df_normalized['age_group'] = target_df
    
    df_normalized = df_normalized.dropna()
    return df_normalized

```

### Gradient Boosting

```python
def tonumclass(df_copy):
    df = df_copy.copy()
    df['age_group'] = df['age_group'].apply(lambda x: 0 if (x == 0.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 1 if (x == 20.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 2 if (x == 30.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 3 if (x == 40.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 4 if (x == 50.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 5 if (x == 60.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 6 if (x == 70.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 7 if (x == 80.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: 8 if (x == 90.0) else x)
    df['age_group'] = df['age_group'].apply(lambda x: int(x))
    
    return df
```

```python
# model to create an xgboost model and perform cross validation

# target needs to be defined

def modelfit(alg, X_train, X_test, y_train, y_test, useTrainCV=True, cv_folds=5, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(X_train.values, label=y_train.values)
        # setting up cross validation
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='mlogloss', early_stopping_rounds=early_stopping_rounds)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    # default eval_metric was 'auc'
    # other values I need to look into 'merror' (Multiclass classification error rate) and 'mlogloss' (Multiclass logloss)
    
    alg.fit(X_train, y_train,eval_metric='mlogloss')
        
    return alg
```

```python
# definition of modelfit(xgb_unbalanced, X_train, X_test, y_train, y_test)
# this is the cross validation function for xgboost model

def xgboost_cv(alg, df, target, av):
    # this will return the average f1 and accuracy scores
    kfold = KFold(5, True, 1)
    
    target = 'age_group'
    predictors = [x for x in df.columns if x not in [target]]
    
    start_time = datetime.datetime.now()
    
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    
    print("df shape: " + str(df.shape))
    
    fold_counter = 1
    
    for train, test in kfold.split(df):
        print("Fold #" + str(fold_counter) + " started...")
        train_df = df.iloc[train,:]
        test_df = df.iloc[test,:]
        X_train = train_df[predictors]
        y_train = train_df[target]
        X_test = test_df[predictors]
        y_test = test_df[target]
        
        print("train shape: " + str(train.shape))
        print("test shape: " + str(test.shape))
        
        trained_model = modelfit(alg, X_train, X_test, y_train, y_test)
        
        predictions = trained_model.predict(X_test)
        
        accuracy = metrics.accuracy_score(y_test.values, predictions)
        precision = metrics.precision_score(y_test.values, predictions, average=None)
        recall = metrics.recall_score(y_test.values, predictions, average=None)
        
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)
        
        print("Fold #" + str(fold_counter) + " ended.\n")
        fold_counter += 1
        
    accuracy_scores = np.array(accuracy_scores)
    precision_scores = np.array(precision_scores)
    recall_scores = np.array(recall_scores)
    
    accuracy_mean = np.mean(accuracy_scores)
    precision_mean = np.mean(precision_scores)
    recall_mean = np.mean(recall_scores)
    
    end_time = datetime.datetime.now()
    time_diff = end_time - start_time
    mins, secs = divmod(time_diff.total_seconds(), 60)
    construction_time = str(mins) + "min " + str(secs) + "sec"
    
    return accuracy_mean, precision_mean, recall_mean, construction_time
```
